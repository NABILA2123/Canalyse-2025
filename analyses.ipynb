```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import folium
from folium.plugins import MarkerCluster
```

```python
# Charger les données depuis le fichier CSV
data = pd.read_csv(r"tec\data.csv")
```

```python
data
```

```python
# Répartition des touristes par âge
plt.figure(figsize=(10, 6))
sns.histplot(data=data, x='Age', bins=20, kde=True)
plt.title('Répartition des touristes par âge')
plt.xlabel('Âge')
plt.ylabel('Nombre de touristes')
plt.show()
```

```python
# Répartition des touristes par nationalité
nationality_counts = data['Nationalité'].value_counts()
plt.figure(figsize=(10, 6))
nationality_counts.plot(kind='bar')
plt.title('Répartition des touristes par nationalité')
plt.xlabel('Nationalité')
plt.ylabel('Nombre de touristes')
plt.xticks(rotation=45)
plt.show()
```

```python
# Répartition des touristes par sexe
sex_counts = data['Sexe'].value_counts()
plt.figure(figsize=(6, 6))
sex_counts.plot(kind='pie', autopct='%1.1f%%')
plt.title('Répartition des touristes par sexe')
plt.ylabel('')
plt.show()
```

```python

# Comparaison entre les différents groupes démographiques pour identifier les tendances
age_nationality = data.groupby(['Nationalité', pd.cut(data['Age'], bins=[0, 20, 40, 60, 80, 100])]).size().unstack(fill_value=0)
plt.figure(figsize=(12, 8))
sns.heatmap(age_nationality, cmap='YlGnBu', annot=True, fmt='d')
plt.title('Répartition des touristes par âge et nationalité')
plt.xlabel('Tranches d\'âge')
plt.ylabel('Nationalité')
plt.show()
```
